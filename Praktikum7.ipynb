Import Library

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data penjualan
brands = ['Samsung', 'Iphone', 'Vivo', 'Oppo']
sales_2023 = [500, 700, 300, 400]
sales_2024 = [550, 750, 350, 450]
sales_2025 = [600, 800, 400, 500]
```

```python
y_pos = np.arange(len(brands))

#canvas dan axis grafik
fig1, ax1 = plt.subplots()

#grafik batang horizontal
ax1.set_title('Horizontal Bar Chart - 2023')
ax1.set_xlabel('Jumlah Penjualan')
ax1.set_ylabel('Merk')
ax1.barh(brands, sales_2023)
ax1.legend()
plt.show()


fig2, ax2 = plt.subplots()
ax2.set_title('Horizontal Bar Chart - 2023')
ax2.set_xlabel('Jumlah Penjualan')
ax2.set_ylabel('Merk')

ax2.barh(brands, sales_2023, label='2023')
ax2.barh(brands, sales_2024, left=sales_2023, label='2024')

ax2.legend()
plt.show()
```

```python
y_pos = np.arange(len(brands))
colors = ['gold', 'skyblue', 'coral', 'lightgreen']
fig3, ax3 = plt.subplots()

ax3.barh(brands, sales_2023, color=colors, edgecolor = 'black', height=0.6)
ax3.set_title('Penjualan SnartPhone tahun 2023')
ax3.set_xlabel('Jumlah Penjualan')
ax3.set_ylabel('Merk')
ax3.grid(axis='x', linestyle='--', alpha=0.7)

for i, v in enumerate(sales_2023):
    ax3.text(v + 10, i, str(v), va='center')
plt.show()

#custom stacked horizontal bar chart
fig4, ax4 = plt.subplots()
ax4.barh(brands, sales_2023, color='lightcoral', edgecolor = 'black', label='2023')
ax4.barh(brands, sales_2024, left=sales_2023, color='lightskyblue', edgecolor = 'black', label='2024')

ax4.set_title('Penjualan Smartphone tahun 2023 dan 2024 (Custom)')
ax4.set_xlabel('Jumlah Penjualan')
ax4.set_ylabel('Merk')
ax4.grid(axis='x', linestyle='--', alpha=0.5)

total_sales = np.array(sales_2023) + np.array(sales_2024)

#label nilai
for i, v in enumerate(total_sales):
    ax4.text(v + 10, i, str(v), color='black')
plt.show()
ax4.legend()
```

***Multiple Bar Chart***

```python
y_pos = np.arange(len(brands))

plt.barh(y_pos - 0.2, sales_2023, color='skyblue', height=0.4, label='2023')
plt.barh(y_pos + 0.2, sales_2024, color='salmon', height=0.4, label='2024')
plt.yticks(y_pos, brands)
plt.title('Penjualan Smartphone 2023 vs 2024')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Merek')
plt.legend()
plt.show()

#stacked
y_pos = np.arange(len(brands))

plt.barh(y_pos, sales_2023, color='skyblue', label='2023')
plt.barh(y_pos, sales_2024, left=sales_2023, color='salmon', label='2024')
plt.barh(
    y_pos,
    sales_2025,
    left=np.array(sales_2023) + np.array(sales_2024),
    color='lime',
    label='2025'
)

plt.yticks(y_pos, brands)
plt.title('Penjualan Smartphone 2023-2025 (Multiple Stacked)')
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Merek')
plt.legend()
plt.show()
```
